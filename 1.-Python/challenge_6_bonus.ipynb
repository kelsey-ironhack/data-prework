```python
from random import choice

gestures=["rock","paper","scissors","lizard","spock"]

def roundsfunc ():
    rounds=input("How many times do you want to play? Please write an odd number ")
    while (int(rounds) % 2) == 0:
        rounds=input("Wrong input. Please write an odd number ")
    else:
        print("Right input. You will play ", rounds, "rounds")
    return rounds

n_rounds=int(roundsfunc())

cpu_score=0
player_score=0

def c_choice():
    computer_choice=choice(gestures) 
    print("The computer chose randomly",computer_choice)
    return computer_choice
    
def p_choice():
    print("Please choose rock, paper, scissors, lizard or spock")
    gestures=["rock","paper","scissors","lizard","spock"]
    choose_type=""
    while choose_type not in gestures:
        choose_type=input()
        if choose_type in gestures:
            print("Right input. You chose", choose_type)
        else:
            print("Wrong input. Please type again")
    return choose_type

def who_wins(p,c):
    
    if (p=="scissors" and c=="rock") or (p=="paper" and c=="scissors") or (p=="rock" and c=="paper") or (p=="lizard" and c=="rock") or (p=="spock" and c=="lizard") or (p=="scissors" and c=="spock") or (p=="lizard" and c=="scissors") or (p=="paper" and c=="lizard") or (p=="spock" and c=="paper") or (p=="rock" and c=="spock"):
        return 1
    elif (p=="rock" and c=="scissors") or (p=="scissors" and c=="paper") or (p=="paper" and c=="rock") or (p=="rock" and c=="lizard") or (p=="lizard" and c=="spock") or (p=="spock" and c=="scissors") or (p=="scissors" and c=="lizard") or (p=="lizard" and c=="paper") or (p=="paper" and c=="spock") or (p=="spock" and c=="rock"):
        return 2
    else:
        return 0
    
counter=0

while counter <n_rounds:
    counter+=1
    print ("Welcome to the round number", counter, "!")
    people_choice=p_choice()
    computer_random_choice=c_choice()
    winner= who_wins(people_choice,computer_random_choice)
    print("If the computer won, 1 is displayed; if the player won, 2 is displayed; and if it was a tie, 0 will be displayed:", winner)
    if winner==1:
        cpu_score+=1
    elif winner==2:
        player_score+=1
    else:
        cpu_score+=1
        player_score+=1
else:
    print ("The game is over. The score is ", cpu_score, "for the computer and", player_score, "for the player")
    if cpu_score>player_score:
        print("The winner is the computer. Better luck next time!")
    elif player_score>cpu_score:
        print("The winner is the player. Congrats!")
    else:
        print("It was a tie. Maybe play again? :)")
```
