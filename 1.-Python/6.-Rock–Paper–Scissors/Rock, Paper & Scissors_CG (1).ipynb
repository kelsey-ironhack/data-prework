```python
import random
```

```python
gestures = ['rock', 'paper', 'scissors']
```

```python
i = 0
n_rounds = 2*i + 1
```

```python
rounds_to_win = 0
```

```python
cpu_score = 0
player_score = 0
```

```python
random.choice(gestures)
```

```python
player_gesture = input("Enter your gesture : ")
while player_gesture not in ['rock', 'paper', 'scissors']:
    player_gesture = input("Enter your gesture : ")
```

```python
'rock' > 'scissors' == True
'paper' > 'rock' == True
'scissors' > 'paper' == True

while rounds_to_win < n_rounds:
    if player_gesture > random.choice(gestures):
        player_score += 2
        print("Player gesture is",player_gesture)
        print("Computer gesture is",random.choice(gestures))
        
    elif player_gesture < random.choice(gestures):
        cpu_score += 1
        print("Player gesture is",player_gesture)
        print("Computer gesture is",random.choice(gestures))
        
    else:
        print("Tie")
        player_score += 0
        cpu_score += 0
```

```python
if cpu_score > player_score:
    print("Computer wins.")
        
elif cpu_score < player_score:
    print("Player wins.")
        
else:
    print("Tie")
```
