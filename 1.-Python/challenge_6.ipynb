```python
#1

from random import choice

#2

gestures=["rock","paper","scissors"]


#3

n_rounds=int(input("How many times you want to play? Please write a number like 1, 3, etc "))

while (n_rounds % 2) == 0:
    n_rounds+=1
    
print("Note: rock, paper, scissors has odd number of rounds, therefore, according to the rules, the game will have", n_rounds, "rounds")

#4

from math import ceil

rounds_to_win=ceil(n_rounds/2)
print("Player must win at least",rounds_to_win, "rounds to win the game. A tie will give a win/point to each player")

#5

cpu_score=0
player_score=0

#6

def c_choice():
    computer_choice=choice(gestures) 
    print("The computer chose randomly",computer_choice)
    return computer_choice

    
#7
    
def p_choice():
    print("Please choose rock, paper or scissors")
    gestures=["rock","paper","scissors"]
    choose_type=""
    while choose_type not in gestures:
        choose_type=input()
        if choose_type in gestures:
            print("Right input. You chose", choose_type)
        else:
            print("Wrong input. Please type again")
    return choose_type

#8 

def who_wins(p,c):
    
    if (p=="scissors" and c=="rock") or (p=="paper" and c=="scissors") or (p=="rock" and c=="paper"):
        return 1
    elif (p=="rock" and c=="scissors") or (p=="scissors" and c=="paper") or (p=="paper" and c=="rock"):
        return 2
    else:
        return 0
    

#9 to 11     
    
counter=0

while counter <n_rounds:
    counter+=1
    print ("Welcome to the round number", counter, "!")
    people_choice=p_choice()
    computer_random_choice=c_choice()
    winner= who_wins(people_choice,computer_random_choice)
    print("If the computer won, 1 is displayed; if the player won, 2 is displayed; and if it was a tie, 0 will be displayed:", winner)
    if winner==1:
        cpu_score+=1
    elif winner==2:
        player_score+=1
    else:
        cpu_score+=1
        player_score+=1
else:
    print ("The game is over. The score is ", cpu_score, "for the computer and", player_score, "for the player")
    if cpu_score>player_score:
        print("The winner is the computer. Better luck next time!")
    elif player_score>cpu_score:
        print("The winner is the player. Congrats!")
    else:
        print("It was a tie. Maybe play again? :)")
```
